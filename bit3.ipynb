```python
def time_ini():
    return datetime.now()
def time_fin(i):
    tim = datetime.now() - i
    return 'Time: {}'.format(tim).split('.')[0]
```

### Descargar la base (rdp_species_assignment_18.fa.gz) de datos 16S a partir de RDP taxonomic:

### Descomprimir el archivo .gz

### Construir una base de datos

```python
subprocess.call('makeblastdb -in db/rdp_species_assignment_18.fa -dbtype nucl -parse_seqids -out db/species_16S', shell = True)
```

### Correr Blastn localmente

```python
init = time_ini()
#---------------
subprocess.call('blastn -db db/species_16S -query bacterias.txt -evalue 1E-6 -outfmt "6 qacc sacc qlen slen length qstart qend sstart send score bitscore evalue pident nident mismatch positive gaps gapopen" -max_target_seqs 1000 -max_hsps 1000 -out resultado.txt', shell = True)   
#----------------
time_fin(i = init)
```

```python
"""
Función para abrir archivos fasta
"""
def open_file(file = ''):
    fas = {}
    with open(file) as fq:
        for line in fq:
            line = line.rstrip()
            if '>' in line:
                header = line.replace('>','')
                s = ''
            else:
                s += line
            fas[header] =  s
    return fas
```

```python
fasta = open_file(file = 'DB/rdp_species_assignment_18.fa')
```

```python
nombres = {}
for i in fasta.keys():
    i = i.split(' ')
    nombres[i[0]] = i[1]
```

```python
nombres
```

```python
df['Species'] = [nombres[i] for i in df.sacc]
```
